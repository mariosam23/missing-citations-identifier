```python
import sys
from pathlib import Path

repo_root = Path.cwd().resolve().parent if Path.cwd().name == "demos" else Path.cwd().resolve()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from src.pipeline.pdf_parser import GrobidPDFParser, NougatPDFParser, ParsedPaper
```

```python
papers_dir = repo_root / "papers"
pdf_files = list(papers_dir.glob("*.pdf"))
```

```python
for pdf in pdf_files:
    grobid_parser = GrobidPDFParser(pdf_path=pdf.as_posix())
    print(f"Parsing {pdf.name} with GrobidPDFParser...")
    grobid_parsed_paper = grobid_parser.parse()
    print(f"Title: {grobid_parsed_paper.title}")
    print(f"Abstract: {grobid_parsed_paper.abstract}...")  # Print first 200 chars of abstract
    print(f"Abstract length: {len(grobid_parsed_paper.abstract)} characters")
    print(f"references: {grobid_parsed_paper.references}...") 
    print(f"Number of references extracted: {len(grobid_parsed_paper.references)}")
    print("-" * 80)
    
    nougat_parser = NougatPDFParser(pdf_path=pdf.as_posix())
    print(f"Parsing {pdf.name} with NougatPDFParser...")
    nougat_parsed_paper = nougat_parser.parse()
    print(f"Title: {nougat_parsed_paper.title}")
    print(f"Abstract: {nougat_parsed_paper.abstract}...")  # Print first 200 chars of abstract
    print(f"Abstract length: {len(nougat_parsed_paper.abstract)} characters")
    print(f"references: {nougat_parsed_paper.references}...")
    print(f"Number of references extracted: {len(nougat_parsed_paper.references)}")
    print("=" * 80)
```
