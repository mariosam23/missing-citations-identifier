- **Nodes**: They are normalized "scientific concepts" (noun-chunks and named entities) extracted from the document's abstract using the SciSpaCy Natural Language Processing model. Before becoming a node, they are filtered to remove generic stopwords (e.g., "method", "result") and low-confidence elements, leaving only core entities (like "node classification" or "graph embedding").

- **Edges**: They represent the relationships between these concepts within the text.
  - A weight of 1 connects concepts that simply co-occur in the same sentence.
  - A weight of 2 connects concepts that share a direct syntactic dependency (i.e., they are grammatically structurally linked together, not merely present in the same sentence).
